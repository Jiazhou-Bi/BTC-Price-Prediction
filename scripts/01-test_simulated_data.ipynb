```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_parquet('../data/00-simulated_data/simulated_data_1h.parquet')
df.head()
```

```python
time_range = df['Time'].max() - df['Time'].min()
print(f"Range of Time column: {time_range}")
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Open-Diff'], bins=30, edgecolor='black', alpha=0.7)
plt.title('Histogram of Open-Diff', fontsize=14)
plt.xlabel('Open-Diff', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
